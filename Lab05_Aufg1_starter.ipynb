### Teil 1

* Iterieren Sie die Folge $x_i = x_{i-1} r (1 - x_{i-1})$ für $r = 2.5$ und $x_0 = 0.2$ für $i\leq 50$.
  Speichern Sie alle $x_i$ in einem NumPy-Array. Plotten Sie diese Werte.
* Ändern Sie (händisch) nun $r$ zwischen $r = 2.5$ und $r = 4$ in Schritten von 0.1 und beobachten Sie, wie sich die Folge verhält.
  Erhöhen Sie gegebenenfalls die maximale Iterationszahl.

````python
import numpy as np
import matplotlib.pyplot as plt
r = 3.7
x0 = 0.2
N = 100
xi = np.zeros(N)
for i in range(N):
    xi1 = x0*r*(1-x0)
    x0 = xi1
    xi[i]= xi1
print(xi)
plt.plot(xi)
plt.show()
````
