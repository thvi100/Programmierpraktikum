 f'(x) &\approx  \frac{f(x+h)-f(x)}{h} \, , \\

### Schritt 3
* Plotten Sie die Fehler $\delta$ aller drei Formeln in das gleiche Diagramm und vergleichen Sie die Kurven miteinander.

````python
import math as m
import numpy as np
import matplotlib.pyplot as plt
x = 2
h = np.logspace(-15,-1,100)
def f(x):
    f = np.sin(x)
    return f
f1a = (f(x+h)-f(x))/h
f1b = (f(x+h)-f(x-h))/(2*h)
f1c = (-f(x+2*h)+8*f(x+h)-8*f(x-h)+f(x-2*h))/(12*h)
f1ana = np.cos(x)
dl1= np.abs(f1a-f1ana)
dl2= np.abs(f1b-f1ana)
dl3= np.abs(f1c-f1ana)

plt.plot(h,dl3)
plt.plot(h,dl2)
plt.plot(h,dl1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('h')
plt.ylabel('delta')
plt.legend(['dl3 = formel3', 'dl2 = formel2', 'dl1 = formel1'])
plt.autoscale(enable=True, tight=True)
plt.grid()
plt.show()
````
